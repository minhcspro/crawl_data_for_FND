```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import csv
```

```python
def get_article_links(page_num):
    with requests.Session() as session:
        url = f'https://tinbaihay.net/s/ch%E1%BB%AFa+b%E1%BB%87nh+down/page-{page_num}.htm'
        response = session.get(url)
        if response.status_code == 200:
            soup = BeautifulSoup(response.content, "html.parser")
            article_List = soup.find_all("div", class_="item-news")
            for item in article_List:
                for link in item.find_all('a'):
                    yield link['href']
```

```python
def scrape_article(link):
    with requests.Session() as session:
        response = session.get(link)
        if response.status_code == 200:
            soup = BeautifulSoup(response.content, "html.parser")
            title = soup.find("h1")
            title = title.text.strip() if title is not None else " "
            paragraphs = soup.find_all("div", class_="content-detail")
            article_content = [
                x.text.strip()
                for content in paragraphs
                for x in content.find_all('p')
            ]
            return {
                "Link": link,
                "Title": title,
                "Content": " ".join(article_content),
                "label": "0",
            }
```

```python
def main():
    with open('/content/Chữa bệnh down.csv', 'w', newline='', encoding='utf-8') as csvfile:
        fieldnames = ["Link", "Title", "Content", "label"]
        writer = csv.DictWriter(csvfile, fieldnames=fieldnames, delimiter='\t')
        writer.writeheader()
        for page_num in range(1, 560):
            article_links = get_article_links(page_num)
            for link in article_links:
                try:
                    data = scrape_article(link)
                    if data:
                        writer.writerow(data)
                except:
                    print(f"An error occurred while scraping article {link}")
```

```python
if __name__ == '__main__':
    main()
```
